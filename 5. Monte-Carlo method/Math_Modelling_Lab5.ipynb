```python
def function_a(x):
  return 1 / (x ** 2 + 2 * x + 4)
```

Подынтегральная функция в пункте б


```python
import numpy as np
import math
def function_b(x):
  return (x * math.log(x)) / (8 + x**3)
```

Функция, которая по методу Монте-Карло будет считать интеграл с границами -1 и 0 для заданной функции и количества реализаций для пункта а

```python
import random
def monte_carlo_integration(func, num_samples):
    integral_sum = 0
    for _ in range(num_samples):
        x = random.uniform(-1, 0)
        integral_sum += func(x)

    interval_length = 1  # Длина интервала [-1, 0]
    estimated_integral = (integral_sum / num_samples) * interval_length
    return estimated_integral
```

Функция, которая по методу Монте-Карло будет считать интеграл с границами 0 и бесконечность для заданной функции и количества реализаций для пункта б

```python
import random

def monte_carlo_importance_sampling(func, num_samples):
    integral_sum = 0
    for _ in range(num_samples):
        u = random.uniform(0, 1)
        integral_sum += func(1/u) / u**2

    estimated_integral = integral_sum / num_samples
    return estimated_integral
```

Оценка интегралов из пункта а и б

```python
num_samples = 1000
estimated_result = monte_carlo_integration(function_a, num_samples)
print(f"Оцена интеграла из пункта а методом Монте-Карло: {estimated_result}")
estimated_result = monte_carlo_importance_sampling(function_b, num_samples)
print(f"Оцена интеграла из пункта б методом Монте-Карло: {estimated_result}")
```

Реализация интеграла для пункта в:

```python
def function_c(x, y):
  return (10 * math.log(4 + x) * math.log(5 + y)) / (x**2 + y ** 2 + 1)

def monte_carlo_double_integration_bounded(func, num_samples):
  integral_sum = 0
  for _ in range(num_samples):
    x = random.uniform(-3, 3)
    y = random.uniform(-3, 3)

    while abs(x) + abs(y) >= 3:
      x = random.uniform(-3, 3)
      y = random.uniform(-3, 3)

    integral_sum += func(x, y)

    interval_area = 36
    return integral_sum / num_samples * interval_area
```

Оцена интеграла для пункта в:

```python
num_samples = 1000
estimated_result = monte_carlo_double_integration_bounded(function_c, num_samples)
print(f"Оцена интеграла из пункта а методом Монте-Карло: {estimated_result}")
```

Значение интеграла, вычисленного с помощью интернет-ресурсов:

```python
import numpy.random as random

def monte_carlo_simulation(A, b, num_samples):
    num_variables = len(b)
    solutions = np.zeros((num_samples, num_variables))

    for i in range(num_samples):
        # Генерируем случайный вектор неизвестных
        random_x = random.randn(num_variables)
        # Решаем систему линейных уравнений с использованием сгенерированного вектора
        solution = np.linalg.solve(A, b + random_x)
        solutions[i, :] = solution

    return solutions

# Пример использования:
A = np.array([[0.18, 0.18, 0.54], [0.18, 0.45, 0.27], [0.54, 0.27, 0.09]])
b = np.array([0.5, 0, 0.5])
num_samples = 10000

solutions = monte_carlo_simulation(A, b, num_samples)

# Статистика по решениям
mean_solution = np.mean(solutions, axis=0)
std_solution = np.std(solutions, axis=0)

print("Среднее решение:", mean_solution)
print("Стандартное отклонение решения:", std_solution)

print(f"Реальное решение: {np.linalg.solve(A, b)}")
```
